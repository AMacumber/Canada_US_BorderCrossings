# Time Series Analysis
*Decompose - Test stationarity - Corellation*

## Libraries

```python
#Importing libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
%matplotlib notebook
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

## Data

```python
can_crossings_000 = pd.read_csv('Can_Crossings.csv', index_col = 0)
can_crossings_000.index = pd.to_datetime(can_crossings_000.index)
can_crossings_000.head()
```

## Exploration

**Americans**

```python
can_crossings_000['Americans'].plot()
plt.show()
```

```python
usa_decompose = seasonal_decompose(can_crossings_000['Americans'], freq = 12, model='multiplicative')
usa_decompose.plot()
plt.show()
```

```python
can_crossings_000['Canadians'].plot()
plt.show()
```

```python
can_decompose = seasonal_decompose(can_crossings_000['Canadians'],
                                  freq = 12,
                                  model = 'multiplicative')
can_decompose.plot()
plt.show()
```
